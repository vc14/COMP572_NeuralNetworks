This is a synthetic customer dataset created for practicing Exploratory Data Analysis (EDA), feature engineering, and machine learning (ANN, regression/classification).
It simulates a subscription-based company (like telecom, SaaS, or online retail) and contains messy, real-world–like data with missing values, outliers, rare categories, and text notes.
The target is churned → whether the customer has left (1) or is still active (0).

```python
import pandas as pd
df=pd.read_csv('eda_sample_customers.csv', parse_dates=['signup_date'])
df.head()
```

```python
# Checking, how many rows/columns and what the columns are.
df.shape, df.columns.tolist()
```

```python
# data types + where missing values exist.
df.info()
```

```python
# A fast overview of numeric stats and categorical counts/top category.
df.describe(include='all').T.head(20)
```

```python
# know how many exact duplicate rows you have
duplicate_count = df.duplicated().sum()
duplicate_count
```

```python
# Printing duplicates including the firts occurence
duplicate_rows=df[df.duplicated(keep=False)]
duplicate_rows
```

```python
# Drop duplicates
df=df.drop_duplicates()
duplicate_count = df.duplicated().sum()
duplicate_count
```
